```python
import os
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
from tqdm.notebook import tqdm
```

```python
data_dir = '/content/drive/MyDrive/data/data'
data_list = os.listdir(data_dir)
data_list
```

```python
path_list = []
for data in data_list:
    path_list.append(data_dir + '/' + data + '/Image/')

path_list
```

```python
len(os.listdir(path_list[0])), len(os.listdir(path_list[1])), len(os.listdir(path_list[2]))
```

```python
image_path = {}
for idx, path in enumerate(path_list):
    image_path[path] = os.listdir(path_list[idx])

image_path
```

```python
img = cv2.imread(path_list[0] + image_path[path_list[0]][0], cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
print(path_list[0])
print(image_path[path_list[0]][0])
print(path_list[0] + image_path[path_list[0]][0])
```

```python
for path in tqdm(path_list):
    for i in image_path[path]:
        img_path = path + i

        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        cv2.imwrite(img_path, img)
```
