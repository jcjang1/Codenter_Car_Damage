# Image GrayScale & Jpg to Png

```python
import os
from PIL import Image


path = '' # GrayScale 해줄 파일경로 지정 , 파일경로 지정해줄 때 맨 뒤에 '/' 꼭 붙여주기
          # 폴더 '내부의' 파일들이기 때문에 경로 끝에 '/' 추가
path_list = os.listdir(path) # os.listdir(path) : path 안의 파일들을 리스트로 불러올 수 있다.
for i in path_list: # path_list를 범위로 지정
    dir = path + i  # 경로 내에서 파일들을 하나씩 가져오기 위해 dir지정
    im_gray = Image.open(dir).convert('L') # dir에서 이미지를 불러와 convert('L')을 사용하여 GrayScale 해줌
    im_gray.save(dir)
```

```python
## 함수
src='' # 이미지 경로
lst=os.listdir(src) # for문에서 범위값 지정을 위해 os,listdir을 사용하여 파일들을 리스트로 반환
def grayscale(src, lst):
    for img in lst:
        img_dir=os.path.join(src, img)
        img_gray=Image.open(img_dir).convert("L")
        img_gray.save(img_dir)
```

```python
## Jpg -> Png
src='' # 이미지 경로
lst=os.listdir(src) # for문에서 범위값 지정을 위해 os,listdir을 사용하여 파일들을 리스트로 반환
def jpg2png(src, lst):
    for jpg in lst:
        jpg_dir=os.path.join(src, jpg)
        jpg_gray=Image.open(jpg_dir)
        jpg_gray.save(jpg_dir[:-4]+'.png','png')
```

# Padding & Resizing

```python
import os
import cv2
import numpy as np
from tqdm.notebook import tqdm
from google.colab.patches import cv2_imshow

## img : 패딩과 리사이징 할 이미지
## set_size : 리사이징 할 크기
## ex) padding(CarDamage.png, 512) -> CarDamage.png를 512x512로 리사이징하며 zero padding 시켜줌
def padding(img, set_size):

  try:
    h,w,c = img.shape # img shape ( height, weight, channel ) 를 h,w,c 변수로 저장
  except:
    print('restart')
    raise

  if h<w: # 원본 이미지에서 h<w일때
    new_width = set_size # 리사이즈할 크기를 입력( set_size )
    new_height = int(new_width*(h/w)) # 리사이즈하되 원본의 h,w 비율을 가져가기위한 계산
  else: # 원본 이미지에서 h>w일때
    new_height = set_size
    new_width = int(new_height*(w/h))

  if max(h,w) < set_size:
    img = cv2.resize(img, (new_width, new_height), cv2.INTER_CUBIC) # INTER_CUBIC : 사이즈를 크게할 떄 사용하는 보간법
  else:
    img = cv2.resize(img, (new_width, new_height), cv2.INTER_AREA) # INTER_AREA : 사이즈를 작게할 때 사용하는 보간법

  try:
    h,w,c = img.shape
  except:
    print('restart')
    raise
  
  delta_w = set_size - w
  delta_h = set_size - h
  top, bottom = delta_h//2,delta_h-(delta_h//2)
  left, right = delta_w//2,delta_w-(delta_w//2)

  new_img=cv2.copyMakeBorder(img,top,bottom,left,right,cv2.BORDER_CONSTANT,value=[0,0,0]) # copyMakeBorder : 패딩 작업에서, 이미지에 포토프레임 같은 경계선을 만들 떄
                                                                                          # BORDER_CONSTANT :  단색의 경계를 추가하는 것, value에서 색의 지정
  return new_img
```
